# Introduction to RooFit

Design: Each mathematical symbol is defined as separate C++ object

#### For example: Gaussian or Normal 1D distribution

Contains \
x: observable = RooRealVar \
μ: mean = RooRealVar \
σ: standard deviation = RooRealVar

```python
import ROOT
m = ROOT.RooRealVar("m", "Bs mass", 5.2, 5.3)
mean = ROOT.RooRealVar("mean", "Mean", 5.28, 5.2, 5.3)
sigma = ROOT.RooRealVar("sigma", "Sigma", 0.0027, 0.001, 1)
signal = ROOT.RooGaussian("signal", "signal", m, mean, sigma)
```

```python
%jsroot off

# Defining a frame to put graphical objects
frame = m.frame()
# Dump parameters to the TPad
signal.plotOn(frame)
frame.Draw()
# Require for Jupyter notebooks only. 
ROOT.gPad.Draw()
```
