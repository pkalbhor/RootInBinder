### Toy study
* Objective is to test reliability of the fitter
  * It is done by producing thousands of toy sample and testing the final PDF over each sample
  * Distribution of final observables should be populated around true value of observables
* It can be used for determining statistical errors when parameter space involves boundary
* Toys are simulated data produced by providing fixed values of the PDF
